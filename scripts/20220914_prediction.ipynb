### Prediction
Model from training LSTM on CoBot/Formica (dump_from_simulation_1)

Inference on other CoBot runs


```python
import os
from antelope_ml.inference_utilities import run_prediction
import json

def parse_params(params_str):
    exp_params = dict()
    for item in os.path.basename(experiment_path).split(","):
        k, v = item.split("=")
        exp_params[k] = v
    return exp_params

trained_on_channel = "dump_from_simulation_1"
run_on_channel = "dump_from_simulation_3"
```

```python
experiment_path = "model=LSTM,layers=2,forecast=10,input_length=150,dataset=formica20220104"
model_path = os.path.join("c:\\experiments", experiment_path, trained_on_channel, "model.pt")
exp_params = parse_params(experiment_path)


metrics = run_prediction(
    model_path,
    input_steps=int(exp_params["input_length"]),
    output_steps=int(exp_params["forecast"]),
    dataset_name=exp_params["dataset"],
    channel_name=run_on_channel,
    batch_size=64,
)
print(json.dumps(metrics, indent=4))
```
